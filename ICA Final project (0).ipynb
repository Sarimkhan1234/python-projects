##  Handling Missing Values

### 3.Count missing/null values

##### This code checks each column for missing (null) values to identify what needs cleaning.

```python
print("\nMissing values per column:\n")
print(data.isnull().sum())
```

.isnull().sum() here provides us the total count of the missing values in each column of our dataset

```python
# Replace string-based errors in 'Age' with NaN
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

 This line of code converts the 'Age' column in the data DataFrame to numeric values, replacing any non-numeric entries with NaN. This is useful for ensuring that the 'Age' column contains only numeric data, which can be important for analysis and calculations.

```python
# Clean 'Smoking_Status' where numeric values were mistakenly inserted
data['Smoking_Status'] = data['Smoking_Status'].replace({0: np.nan, 1: np.nan})
```

This line of code replaces all occurrences of 0 and 1 in the 'Smoking_Status' column with np.nan. This can be useful if you want to mark these values as missing or invalid for further analysis.

```python
# Replace empty strings with NaN in specific columns
data['Occupation_Exposure'].replace('', np.nan)
data['Education_Level'].replace('', np.nan)
data['Alcohol_Consumption'].replace('', np.nan)
data['Lung_Cancer_Stage'].replace('', np.nan)
```

 Each line of code is replacing empty strings in specific columns with np.nan, which is useful for data cleaning and ensuring that missing or invalid data is properly marked.

```python
# Final check of missing values after all replacements
print("\nUpdated missing values after cleaning:\n")
print(data.isnull().sum())
```

Finally checking once more that missing values exist or not in the dataframe for a quick verification

### 4. Impute missing values

##### Impute Missing Values refers to the process of replacing missing or incomplete data in a dataset with estimated values. This technique helps to maintain the integrity of the dataset and allows for more accurate analysis and modeling. Common methods include using the mean, median, mode, or more advanced approaches like regression or machine learning algorithms to predict the missing entries.


```python
# Fill missing numeric values with mean
data['Age'].fillna(data['Age'].mean())
data['BMI'].fillna(data['BMI'].mean())
data['Alcohol_Consumption'].fillna(data['Alcohol_Consumption'].mode()[0])
data['Lung_Cancer_Stage'].fillna(data['Lung_Cancer_Stage'].mode()[0])
```

Here we are basically filling all the missing values in our dataframe with the respective averages i.e mean, median, mode column

```python
#converting data['Age'] (float) to integer
data['Age'] = data['Age'].round().astype(int)
```

Here we are converting the age column of our dataset from float datatype to integer datatype

```python
# Fill missing categorical values with mode
for col in ['Smoking_Status', 'Occupation_Exposure', 'Education_Level']:
    data[col].fillna(data[col].mode()[0])
```

Here we are filling in the missing values of a column with its mode computed

### 5. Fix inconsistent data in categorical columns

##### Fix Inconsistent Data in Categorical Columns involves identifying and correcting variations or errors in category labels within a dataset. This includes standardizing different spellings, capitalizations, or formats of the same category (e.g., “Male,” “male,” and “M” all representing the same group). Cleaning these inconsistencies ensures accurate grouping, analysis, and modeling of categorical data.


```python
# Clean 'Gender' where 1=Male, 2=Female were inconsistently mixed
data['Gender'] = data['Gender'].replace({1: 'Male', 2: 'Female'})
data['Gender'] = data['Gender'].str.capitalize()
```

The first line of code replaces numeric values 1 and 2 in the 'Gender' column with the strings 'Male' and 'Female', respectively. The second line of code then capitalizes the first letter of each string in the 'Gender' column to ensure proper formatting.

```python
# Fix case inconsistencies in 'Smoking_Status'
data['Smoking_Status'] = data['Smoking_Status'].str.capitalize()
```

Another way to capitalize or make the string data in the smoking status column uppercase

### 6. Validate cleanup

##### Validate Cleanup is the process of verifying that all data cleaning steps, such as handling missing values and fixing inconsistencies, have been correctly applied. This involves checking the dataset to ensure no errors remain, the data is consistent, and it is ready for accurate analysis or modeling. Validation helps maintain data quality and reliability throughout the workflow.
